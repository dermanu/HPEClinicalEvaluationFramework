```python
import os

def replace_in_files(root_dir, old_str, new_str):
    for subdir, _, files in os.walk(root_dir):
        for file in files:
            if file.endswith(".trc"):
                filepath = os.path.join(subdir, file)
                with open(filepath, 'r') as f:
                    filedata = f.read()
                # Replace the old string with the new one
                newdata = filedata.replace(old_str, new_str)
                with open(filepath, 'w') as f:
                    f.write(newdata)

# Replace 'inf.0' with '100.0' in all .trc files within specified folders
root_directory = '/media/emanu/LaCie/2_VICON_DATA_COMPLETE/Done'  # Update with the actual parent directory
old_string = 'inf.0'
new_string = '100.0'
replace_in_files(root_directory, old_string, new_string)
```
